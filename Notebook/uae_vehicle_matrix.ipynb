# 🇦🇪 UAE Vehicle Matrix Analysis

This notebook loads, explores, and analyzes the **UAE Vehicle Matrix** dataset containing vehicle models, pricing, technical specifications, and classifications.

### 🛠️ 1. Setup & Imports

To use the `pandas` library (usually abbreviated as `pd`), we must import it first. This cell resolves the `Could not find name pd` error.

```python
import pandas as pd
import numpy as np

print("Pandas imported successfully!")
```

### 📊 2. Load the Dataset

We load the Excel matrix sheet from the `Data/` folder. Since this notebook is located in the `Notebook/` subdirectory, we use a relative path `../Data/uae_vehicle_matrix.xlsx`.

```python
# Load the vehicle matrix excel file
file_path = "../Data/uae_vehicle_matrix.xlsx"
df = pd.read_excel(file_path)

# Display the first 5 rows
df.head()
```

### 🔍 3. Basic Exploration & Overview

Let's check the size of the dataset, columns, and data types.

```python
print(f"Dataset contains {df.shape[0]} rows and {df.shape[1]} columns.\n")
df.info()
```

### 💰 4. Pricing & Specification Summary

Let's see statistical summaries for the pricing and other continuous columns.

```python
# Statistical summary of the pricing columns
df[['Min Price (AED)', 'Max Price (AED)', 'Avg Price (AED)', 'Power (HP)', 'Engine Size']].describe()
```

### 🔋 5. Electric Vehicles (EVs) Analysis

Let's see the count and proportion of Electric Vehicles in the matrix.

```python
if 'EV Y/N' in df.columns:
    ev_counts = df['EV Y/N'].value_counts(dropna=False)
    print("Electric Vehicle counts:")
    print(ev_counts)
else:
    print("Column 'EV Y/N' not found.")
```

### 🏢 6. Top Manufacturers

Let's count how many models each manufacturer (`Make`) has in the matrix.

```python
top_makes = df['Make'].value_counts().head(15)
print("Top 15 Manufacturers in the dataset:")
print(top_makes)
```
